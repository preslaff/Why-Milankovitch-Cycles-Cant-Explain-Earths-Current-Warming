```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range for the plot (in years)
time = np.linspace(0, 200000, 1000)  # 200,000 years with 1000 points

# Define parameters for the cycles
# Eccentricity parameters
eccentricity_mean = 0.03
eccentricity_amplitude = 0.02
eccentricity_period_1 = 100000  # 100,000 years
eccentricity_period_2 = 400000  # 400,000 years

# Obliquity parameters
obliquity_mean = 23.5
obliquity_amplitude = 1.3
obliquity_period = 41000  # 41,000 years

# Precession parameters
precession_mean = 0
precession_amplitude = 23.0
precession_period = 26000  # 26,000 years

# Calculate the cycles
eccentricity = (eccentricity_mean +
                eccentricity_amplitude * np.cos(2 * np.pi * time / eccentricity_period_1) +
                (eccentricity_amplitude / 2) * np.cos(2 * np.pi * time / eccentricity_period_2))

obliquity = (obliquity_mean +
             obliquity_amplitude * np.cos(2 * np.pi * time / obliquity_period))

precession = (precession_mean +
              precession_amplitude * np.cos(2 * np.pi * time / precession_period))

# Combined effect (for illustration purposes, simply sum the normalized effects)
combined_effect = (eccentricity / max(eccentricity) +
                   obliquity / max(obliquity) +
                   precession / max(precession))

# Plotting the results
plt.figure(figsize=(14, 10))

# Plot eccentricity
plt.subplot(4, 1, 1)
plt.plot(time, eccentricity, label='Eccentricity', color='b')
plt.xlabel('Time (years)')
plt.ylabel('Eccentricity')
plt.legend()

# Plot obliquity
plt.subplot(4, 1, 2)
plt.plot(time, obliquity, label='Obliquity', color='r')
plt.xlabel('Time (years)')
plt.ylabel('Axial Tilt (degrees)')
plt.legend()

# Plot precession
plt.subplot(4, 1, 3)
plt.plot(time, precession, label='Precession', color='g')
plt.xlabel('Time (years)')
plt.ylabel('Precession')
plt.legend()

# Plot combined effect
plt.subplot(4, 1, 4)
plt.plot(time, combined_effect, label='Combined Effect', color='m')
plt.xlabel('Time (years)')
plt.ylabel('Normalized Combined Effect')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range for the plot (in years)
time = np.linspace(0, 200000, 1000)  # 200,000 years with 1000 points

# Define parameters for the cycles
# Eccentricity parameters
eccentricity_mean = 0.03
eccentricity_amplitude = 0.02
eccentricity_period_1 = 100000  # 100,000 years
eccentricity_period_2 = 400000  # 400,000 years

# Obliquity parameters
obliquity_mean = 23.5
obliquity_amplitude = 1.3
obliquity_period = 41000  # 41,000 years

# Precession parameters
precession_mean = 0
precession_amplitude = 23.0
precession_period = 26000  # 26,000 years

# Calculate the cycles
eccentricity = (eccentricity_mean +
                eccentricity_amplitude * np.cos(2 * np.pi * time / eccentricity_period_1) +
                (eccentricity_amplitude / 2) * np.cos(2 * np.pi * time / eccentricity_period_2))

obliquity = (obliquity_mean +
             obliquity_amplitude * np.cos(2 * np.pi * time / obliquity_period))

precession = (precession_mean +
              precession_amplitude * np.cos(2 * np.pi * time / precession_period))

# Realistic temperature change model (simplified for illustration)
# Use normalized cycles to approximate temperature changes
# Assume a temperature amplitude of +/- 5C for simplicity
temperature_change = 5 * (eccentricity / max(eccentricity) +
                          obliquity / max(obliquity) +
                          precession / max(precession)) - 7.5  # Center around a mean

# Plotting the results
plt.figure(figsize=(14, 8))

# Plot global temperature change
plt.plot(time, temperature_change, label='Global Temperature Change', color='c')
plt.xlabel('Time (years)')
plt.ylabel('Temperature Change (C)')
plt.title('Global Temperature Change Over 200,000 Years Due to Milankovitch Cycles')
plt.legend()

plt.tight_layout()
plt.show()
```
