```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft

# Simulate some real temperature anomaly data (replace this with actual data)
time = np.linspace(1880, 2023, 140)
temperature_anomalies = 0.5 * np.sin(2 * np.pi * time / 70) + 0.2 * np.sin(2 * np.pi * time / 30) + np.random.normal(0, 0.1, len(time))

# Perform Fourier Transformation
temperature_anomalies_fft = fft(temperature_anomalies)

# Frequencies corresponding to the FFT components
frequencies = np.fft.fftfreq(len(temperature_anomalies), d=1)

# Plot the original temperature anomalies
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(time, temperature_anomalies, label='Temperature Anomalies')
plt.xlabel('Time (years)')
plt.ylabel('Temperature Anomaly (°C)')
plt.title('Temperature Anomalies')
plt.legend()

# Plot the magnitude of the Fourier coefficients
plt.subplot(2, 1, 2)
plt.stem(frequencies, np.abs(temperature_anomalies_fft), 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Frequency (1/years)')
plt.ylabel('Magnitude')
plt.title('Fourier Transform of Temperature Anomalies')
plt.xlim(0, 0.1)  # Limiting x-axis for better visibility of low frequencies
plt.legend()

plt.tight_layout()
plt.show()

# Identify frequencies corresponding to Milankovitch cycles
eccentricity_freq = 1 / 100000
obliquity_freq = 1 / 41000
precession_freq = 1 / 26000

# Find the indices of these frequencies in the FFT result
eccentricity_index = np.argmin(np.abs(frequencies - eccentricity_freq))
obliquity_index = np.argmin(np.abs(frequencies - obliquity_freq))
precession_index = np.argmin(np.abs(frequencies - precession_freq))

# Reconstruct the signal using only these frequencies (optional)
reconstructed_fft = np.zeros_like(temperature_anomalies_fft)
reconstructed_fft[eccentricity_index] = temperature_anomalies_fft[eccentricity_index]
reconstructed_fft[obliquity_index] = temperature_anomalies_fft[obliquity_index]
reconstructed_fft[precession_index] = temperature_anomalies_fft[precession_index]

reconstructed_signal = ifft(reconstructed_fft)

# Plot the reconstructed signal
plt.figure(figsize=(12, 6))
plt.plot(time, temperature_anomalies, label='Original Temperature Anomalies')
plt.plot(time, np.real(reconstructed_signal), label='Reconstructed Signal')
plt.xlabel('Time (years)')
plt.ylabel('Temperature Anomaly (°C)')
plt.title('Original and Reconstructed Temperature Anomalies')
plt.legend()
plt.show()
```
